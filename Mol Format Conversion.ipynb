```python
def convertMolFormats(mol_str, input_format='smiles', output_format='sdf'):    
    obConversionOut = ob.OBConversion()
    obConversionOut.SetInAndOutFormats(input_format, output_format)

    molecule = ob.OBMol()
    obConversionOut.ReadString(molecule, mol_str)
    
    return obConversionOut.WriteString(molecule)
```

```python
fm = convertMolFormats("C(=O)([N+](=O)[O-])O")
print(fm)
```

```python
fm2 = convertMolFormats(fm, 'sdf', 'smiles')
print(fm2)
```

```python
import pybel

def xyz_to_smiles(fname: str) -> str:
    
    mol = next(pybel.readfile("xyz", fname))

    smi = mol.write(format="smi")

    return smi.split()[0].strip()
```
